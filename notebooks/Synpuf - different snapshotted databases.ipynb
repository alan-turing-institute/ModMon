Driver obtained on Mac with the following: https://docs.microsoft.com/en-us/sql/connect/odbc/linux-mac/install-microsoft-odbc-driver-sql-server-macos?view=sql-server-ver15#microsoft-odbc-17

```python
import pandas as pd
import pyodbc
driver = "/usr/local/lib/libmsodbcsql.17.dylib"
cnxn_string = 'Driver={' + driver + '};Server=51.104.224.106,1433;Database=WEEK_X;UID=analysts;PWD=An8lysts.'
```

```python
def get_cnxn(n):
    c_str = cnxn_string.replace("X", "{0:0=2d}".format(n))
    return pyodbc.connect(c_str)
```

```python
pd.read_sql('SELECT * FROM person;', get_cnxn(1)).head()
```

```python
pd.read_sql('SELECT * FROM person;', get_cnxn(10)).head()
```

```python
pd.read_sql("SELECT COUNT(person_id) AS person_count FROM person", get_cnxn(1))["person_count"][0]
```
